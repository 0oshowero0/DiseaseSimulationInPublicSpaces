```python
import numpy as np
import json, pickle, importlib, math, random
from pathlib import Path
import pandas as pd
from datetime import datetime
from tqdm import tqdm

scenario='airport'
method="social_force"
control=[]
begin_time=1473638400
end_time=1473638400 + 3600
period=1
folder_path='../../data'
heatmap=None

from SocialForceSimulator import SocialForceSimulator
import Engine2 as Engine
import toml
"""
params: 
    methods: 'social_force' or 'gan'; 
    control: ['social_distancing', 'more_security_ck', 'speed_up', 'reduce_explore']
    begin_time: unixtime
    end_time: unixtime
    period: int, default 1
return: json
{id:{'trajectory': [[x0, y0], [x1, y1], ...], 'state': {}}}
"""

control.sort()
file_name = '{0}/{1}/trace/{2}-{3}-{4}-{5}-{6}.json'.format(
    folder_path, scenario, method, begin_time, end_time, '-'.join(control), period)
            
##### Processing Control Method #####
config = toml.load('../default.toml')
            
if 'speed_up' in control:
    print('Pedestrians speed up')
    config['Pedestrian']['max_speed_multiplier'] *= 2
if 'social_distancing' in control:
    print('Applying Social Distancing')
    config['PedRepulsiveForce'] = config['PedRepulsiveForceSocialDistancing']
if 'reduce_explore' in control:
    print('Explore Reduced')
    config['explore_rate'] *= 0.3
if 'more_security_ck' in control:
    print('Open More Security Check')

config['more_security_ck'] = ('more_security_ck' in control)
print(config)
            
def timestamp2str(timestamp):
    return datetime.strftime(datetime.fromtimestamp(timestamp), "%Y/%m/%d %H:%M:%S")

def str2timestamp(string):
    return int(datetime.strptime(string, "%Y/%m/%d %H:%M:%S").timestamp())

data_path = Engine.file_path(folder_path)
if type(heatmap) == type(None):
    heatmap = np.load(data_path['distribution'])
str_begin, str_end = timestamp2str(begin_time), timestamp2str(end_time)
obstacles = np.array(Engine.load_airport_lines(data_path, more_security_ck=config['more_security_ck'], vertical_flip=True))
```

```python
file_name = '{0}/{1}/trace/{2}-{3}-{4}-{5}-{6}.json'.format(
                folder_path, scenario, method, begin_time, end_time, '-'.join(control), period)
```

```python
from matplotlib import pyplot as plt
from tqdm import tqdm

def static_plot(dataframe, obstacles):
    fig = plt.figure(figsize=(10,10))
    fig.set_tight_layout(True)
    ax = fig.add_subplot(111)
    ax.set_aspect(1.0, 'datalim')
    plt.axis('off')
    
    for iter in obstacles:
        ax.plot(np.array(iter).reshape([-1, 2])[0], np.array(iter).reshape([-1, 2])[1], color='black')

    for iter in tqdm(range(dataframe.shape[0])):
        ax.plot(
            [x[0] for x in dataframe['waypoint'].iloc[iter]],
            [x[1] for x in dataframe['waypoint'].iloc[iter]],
            'red')
        ax.scatter(dataframe['waypoint'].iloc[iter][0][0], dataframe['waypoint'].iloc[iter][0][1], c='blue', marker='X',linewidths=5)
        
static_plot(df2, obstacles)
```

```python
import torch

df2 = df2.join(pd.DataFrame({
    'id' : df2.index,
    'start_time' : df2['checkin_time'],
    'init_velocity' : np.ones([len(df2.index), 2]).tolist(), 
    'arrive_distance_threshold' : 2 * np.ones([len(df2.index)]),
    'max_speed' : config['Pedestrian']['max_speed_multiplier'] * np.ones([len(df2.index)]),
    'infect_state' : np.zeros(df2.shape[0]).tolist()
}))

s = SocialForceSimulator(config, df2, obstacles, start_time=begin_time, device=torch.device('cuda:0'))
bar = tqdm(np.arange(end_time - begin_time), ncols=60)
for iter in bar:
    bar.set_description('ActivePeds:{0}'.format(len(s.ped.active_peds)))
    s.step()

d = s.ped.to_dict()

with open(file_name, 'w') as f:
    json.dump(d, f)
    print('Trajectory file saved in {0}'.format(file_name))
```
